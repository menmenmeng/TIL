```python
# DataFrame 생성하기

# 1. 1차원 ndarray

col_name = ['col1']
list1 = [1, 2, 3]
array1 = np.array(list1)
df_list1 = pd.DataFrame(list1, columns = col_name)
df_array1 = pd.DataFrame(array1, columns = col_name)

print(df_list1)
print(df_array1)
```

```python
# 2. 2차원 ndarray

col_name = ['col1','col2','col3']
list2 = [[1,2,3],[11,12,13]]
array2 = np.array(list2)
df_list2 = pd.DataFrame(list2, columns = col_name)
df_array2 = pd.DataFrame(array2, columns = col_name)

print(df_list2)
print(df_array2)
```

```python
# 3. Dictionary로 DataFrame 만들기
# Dict로 Df를 만들 때는 key 하나에 할당된 데이터가 하나의 "열"을 말하게 된다.

dict1 = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict1 = pd.DataFrame(dict1)
print(df_dict1)


# 4. Series로 DataFrame (Dict의 Key-value 관계와 Series의 index-value 관계가 비슷해 보이지만,
#    Dict의 key는 열 하나를 뜻하는데 Series의 index는 그냥 행 index가 됨.)
#    1차원 df밖에 만들 수 없음

series1 = pd.Series({'1':[1, 11], '2':[2, 22], '3':[3, 33]})
df_series1 = pd.DataFrame(series1)
print(df_series1)
```

```python
# *** DataFrame을 ndarray, list, dictionary로 변환하기 *** 
# 1. ndarray로 변환 : df.values

print(df_dict1.values, type(df_dict1.values), '\n')

# 2. list로 변환 : tolist()

print(df_dict1.values.tolist(), type(df_dict1.values.tolist()), '\n')

# 3. dictionary로 변환 : df.to_dict()

print(df_dict1.to_dict(), type(df_dict1.to_dict()), '\n')
# to_dict() 인자로 아무것도 집어넣지 않으면 각 key값의 value도 dict로 호출
# list로 호출하게금 하려면 to_dict('list')해주기

print(df_dict1.to_dict('list'), type(df_dict1.to_dict('list')))
```

```python
## 다시 titanic_df에서.

# 새 column 생성하기 : 새 column 이름을 인덱싱하듯이 해서 아무 값이나 기입하면 됨.

titanic_df['Age_0'] = 0
titanic_df.head(3)
```

```python
# groupby() : 처음 보는 메소드. // 어떤 컬럼에 대해서. 그 컬럼 안에 있는 value들을. 
# 같은 것끼리 Group지은 다음에, 각 그룹마다 Aggregation 함수를 적용 -> 합쳐 주는 것.

titanic_groupby = titanic_df.groupby('Pclass').count()
# Pclass 열에 대해서. 그 열의 데이터값들 중 동일한 것들끼리 그룹지은 다음에.
# 각 데이터값 그룹마다 count()를 적용해서
# 그걸 합치는 것

titanic_groupby
```

```python
# 객체에 PassengerId, Survived 컬럼만 count 수행하기

titanic_groupby = titanic_df.groupby('Pclass')[['PassengerId', 'Survived']].count()
titanic_groupby
```

```python
# 객체에 한 컬럼을 다른 agg함수로 여러 개 넣는 방법

titanic_df.groupby('Pclass')['Age'].agg([max, min])
```

```python
# 객체에 그냥 여러 포맷 사용. Age는 max, Sibsp는 sum, Fare는 mean을 groupby하고 싶다.

agg_format = {'Age':'max', 'SibSp':'sum', 'Fare':'mean'}
titanic_df.groupby('Pclass').agg(agg_format)
```

```python
# 결손 데이터 처리하기

# 1-1. 결손 데이터 확인 - True or False로
titanic_df.isna().head(3)
```

```python
# 1-2. 결손 데이터 확인 - 몇 개인지
titanic_df.isna().sum()
```

```python
# 결손 데이터 처리하기
# titanic_df['Cabin']의 na값을 C000으로 채운 것을, titanic_df['Cabin']에 새롭게 넣어준다.

titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df.head(3)

# 반환 값을 다시 받아야 하는 메소드임, 주의할 것.
```

```python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna('S')
titanic_df.isna().sum()

titanic_df = titanic_df.fillna(method = 'ffill')
titanic_df.head(10)
```

```python
# apply lambda 데이터 가공

lambda_square = lambda x: x**2
a = [1, 2, 3]

# map 이용하면 한번에 여러 개 인자 동시에 받을 수 있다.
square_a = map(lambda_square, a) # map(function, iteration_data_structure)
list(square_a)
```

```python
# 새로운 컬럼 'Name_len' : 'Name'의 길이

titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x: len(x)) # apply 안에 function을 넣으면 됨
titanic_df.head(3)
```

```python
# lambda 에 if else 절을 이용해 나이에 따라 Child, Adult 구분하는 컬럼 만들기

titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x: 'Child' if x<15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(10)

# 중요 : lambda 안에서 elif 절을 쓸 수 없다. 그러므로 만드려면 그냥 def로 함수를 만들어서 넣는 게 편할 듯
```

```python
# Child, Adult, Elderly로 구분하기

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: 'Child' if x<15 else('Adult' if x<=60
                                                                              else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```
