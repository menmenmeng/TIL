## 0. Generate example data.

```python
# Generate example data.
ex_data = {
    'born' : [1984, 1992, 1988], 
    'name' : ['John', 'May', 'Frank'], 
    'Job':['DataScientist', 'Nurse', 'Farmer']
}

df = pd.DataFrame(ex_data)

print('ex_data : ')
display(df)
```

## 1. Save df as pickle, and read .pkl file.

```python
# Save df as pickle, and read .pkl file.
df.to_pickle('ex_data.pkl')
df1 = pd.read_pickle('ex_data.pkl')

print('Load ex_data from pickle')
display(df1)
```

## 2. Save df as table, and read .tbl file

```python
# Save df as table, and read .tbl file
'''
Table file?
 == All general delimited file. Sep is not fixed, so have to check the delimiter of the file first.
'''
df2 = pd.read_table('ex_data.txt', sep='\s+') # for all spaces.
display(df2)
df2_2 = pd.read_table('ex_data.txt', sep='\t') # why is this not working?
display(df2_2)
```

## 3. csv file

```python
# Second import part of reading, saving dataframe instances.

# Default : index column will be stated clearly in .csv file.
df.to_csv('ex_data.csv')
df3 = pd.read_csv('ex_data.csv')
print('to params : index=True') ; display(df3)

# If index column is not needed(ex. The index increases by 1 from 0.).
df.to_csv('ex_data.csv', index=False)
df3_2 = pd.read_csv('ex_data.csv')
print('to params : index=False') ; display(df3_2)

# If don't want to use index_column, index_col param in read_csv method should be 0.(first column)
df.to_csv('ex_data.csv')
df3_3 = pd.read_csv('ex_data.csv', index_col=0)
print('read params : index_col=0') ; display(df3_3)

# If column names are passed explicitly, then set header parameter as None.
df.to_csv('ex_data.csv', header=None)
df3_4 = pd.read_csv('ex_data.csv', header=None)
print('if header is not needed : ') ; display(df3_4)
```
